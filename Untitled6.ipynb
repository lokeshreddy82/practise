```python
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,z)
```

```python
plt.plot(x,y)
plt.xlabel("angles")
plt.ylabel("sin")
plt.title("this for practise")
plt.show()
```

```python
#we perform double operations
plt.plot(x,y,"r+")
plt.plot(x,z,"g.")
plt.show()
```

```python
#here we perform parabola oprations
x=np.linspace(-12,12,20)
y=x**2
plt.plot(x,y,"r+")
plt.show()
```

```python
fig=plt.figure()
ass=fig.add_axes([0,0,1,1,])
languages=["english","telugu","tamil","kanada","hindhi"]
people=[10,20,30,25,40]
ass.bar(languages,people)
plt.xlabel('languages')
plt.ylabel('Number of people')
plt.show()
```

```python
fig1=plt.figure()
ax=fig1.add_axes([0,0,1,1])
subjects=["englsh","maths","physics","chemistry","biology"]
people=[10,30,50,20,50]
ax.pie(people,labels=subjects)
plt.title("piechart")
plt.show()
```
